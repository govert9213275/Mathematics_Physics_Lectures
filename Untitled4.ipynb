```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate range
def projectile_range(v0, theta_deg, g=9.81, h0=0):
    """Calculate the horizontal range of a projectile.

    Parameters:
    v0 -- initial velocity (m/s)
    theta_deg -- angle of projection in degrees
    g -- gravitational acceleration (m/s^2)
    h0 -- initial height (m)
    """
    theta = np.radians(theta_deg)  # Convert to radians
    term1 = v0 * np.cos(theta)
    term2 = v0 * np.sin(theta)
    term3 = np.sqrt((v0 * np.sin(theta))**2 + 2 * g * h0)
    time = (term2 + term3) / g  # Time of flight
    return term1 * time

# Parameters
v0_values = [10, 20, 30]  # Initial velocities (m/s)
g = 9.81  # Gravity (m/s^2)
h0_values = [0, 10]  # Initial heights (m)
angles = np.linspace(0, 90, 91)  # Angles from 0 to 90 degrees

# Plot range vs angle for different conditions
plt.figure(figsize=(10, 6))
for v0 in v0_values:
    for h0 in h0_values:
        ranges = [projectile_range(v0, theta, g, h0) for theta in angles]
        plt.plot(angles, ranges, label=f'v0 = {v0} m/s, h0 = {h0} m')

# Plot settings
plt.title('Projectile Range vs Angle of Projection')
plt.xlabel('Angle of Projection (degrees)')
plt.ylabel('Range (meters)')
plt.grid(True)
plt.legend()
plt.show()

# Find optimal angle for maximum range
for v0 in v0_values:
    for h0 in h0_values:
        ranges = [projectile_range(v0, theta, g, h0) for theta in angles]
        max_range = max(ranges)
        optimal_angle = angles[np.argmax(ranges)]
        print(f'For v0 = {v0} m/s, h0 = {h0} m: Max range = {max_range:.2f} m at {optimal_angle}Â°')
```
